```python
import os

import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
```

```python
os.chdir('..')
os.getcwd().split('/')[-1] == 'GeneralLoss'
```

```python
from minimizers.minimize import GeneralLossMinimizer
```

```python
seed = 0
```

```python
X, y = make_classification(
    n_classes=2,
    n_samples=1000,
    n_features=10,
    flip_y=0.,
    n_clusters_per_class=1,
    random_state=seed,
)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=seed)
del X, y
```

---

```python
def bce_loss(y, y_hat):
    entropy = (y)*np.log(y_hat) + (1-y)*np.log(1-y_hat)
    return -entropy
def sigmoid_link(X, b):
    y_hat = 1 / (1+np.exp(-X@b.T))
    return y_hat.clip(1e-9, 1-1e-9)
```

```python
model = Pipeline([
    ('sc', StandardScaler()),
    ('est', GeneralLossMinimizer(
        loss_fn=bce_loss,
        link_fn=sigmoid_link,
        random_state=seed,
    )),
])
model[-1].set_estimator_type('classifier')
model[-1].set_multi_output(False)

model.fit(X_train, y_train)
y_hat = (model.predict(X_train)>0.5).astype(int).flatten()
print('Training accuracy:', (y_train==y_hat).mean())

y_hat = (model.predict(X_test)>0.5).astype(int).flatten()
print('Testing accuracy:', (y_test==y_hat).mean())
```
